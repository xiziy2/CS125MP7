```python
%matplotlib inline
from IPython.display import display # inline display function jupyter notebook
import matplotlib.pyplot as plt     # de facto plotting and visualization package in python  
import seaborn as sns               # visualization package based on matplotlib, aesthetic display
import statsmodels.api as sm        # a package with numerous statistical tools
import pandas as pd                 # de facto package for manipulating tabular data
import numpy as np                  # numerical manipulation on sequences of data, basic math and advanced linear algebra
df = pd.read_csv('./Pokemon.csv') # use pandas to read in a csv file as a pd.DataFrame
df.head()                         # display top 5 rows in the data
df = df.drop('#', axis=1)  
yours = pd.read_csv('PokemonYouHave.csv')
```

Display the pokemons you have

```python
yours 
```

Find Enemy pokemon in database:
Example: Raikou

```python
laikou  = df[df['Name']=='Raikou']
display(laikou)
```

Compare with the pokemons you have

```python
enemy = df[df['Name']=='Raikou'].head(50)
plt.scatter(enemy['Attack'], enemy['Defense'], color='r',label='Enemy')
plt.scatter(yours['Attack'], yours['Defense'], color='b',label='Pokemons you have')
plt.legend()
plt.xlabel('Attack')
plt.ylabel('Defense')
```

```python
enemy = df[df['Name']=='Raikou'].head(50)
plt.scatter(enemy['HP'], enemy['Speed'], color='r',label='Enemy')
plt.scatter(yours['HP'], yours['Speed'], color='b',label='Pokemons you have')
plt.legend()
plt.xlabel('HP')
plt.ylabel('Speed')
```
